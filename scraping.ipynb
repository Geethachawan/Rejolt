```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.flipkart.com/clothing-and-accessories/topwear/tshirts/pr?sid=clo,ash,ank&p[]=facets.ideal_for%255B%255D%3DWomen&p[]=facets.ideal_for%255B%255D%3DWomen&otracker=categorytree'
```

```python
req = requests.get(url)
```

```python
content=BeautifulSoup(req.content,'html.parser')
```

```python
data = content.find_all('div',{'class':'_1xHGtK _373qXS'})

```

```python
links=[]
tshirts_name=[]
start_link="https://www.flipkart.com"
```

```python
for items in data:
    rest_link=items.find('a')['href']
    name=items.find('div',attrs={'class':"_2WkVRV"})
    tshirts_name.append(name.text)
    links.append(start_link+rest_link)
    
```

```python
print(tshirts_name[0])
print(links[0])
```

```python
dataframe={'tshirts_names':tshirts_name, 'Links':links}
Final_dataframe=pd.DataFrame(dataframe)
print(Final_dataframe)
```

```python
Final_dataframe.to_csv('fk_data_url.csv')
```
