```python
import matplotlib.pyplot as plt
 

x = ['MATH','HINDI','ENGLISH']
y = [58, 69, 92]
 

plt.plot(x, y)
plt.xlabel('Subjects')
 

plt.ylabel('Marks')
plt.title('Marks Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
sub=['MATH','HINDI','ENGLISH']
S1=[50,68,90]
S2=[50,90,79]
plt.plot(sub,S1,label="Ram")
plt.plot(sub,S2,label="Sham")
plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('marks comparison')
plt.legend()
plt.show()
```
